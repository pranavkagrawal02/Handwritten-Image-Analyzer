<h4 style='color:black' align="left">Train your model for K = 3, 5, 7, 9, 11 or any 5 ‘K’ values using KNN classifier on the following training data:</h4>
<h4 style='color:black' align="left">Training Dataset to be used: train.csv. Test Dataset to be used: test.csv</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
X_train = df_train.drop('label',axis='columns')
y_train = df_train['label']

X_test = df_test.drop('label',axis='columns')
y_test = df_test['label']
```

<h3 style='color:blue' align="left">Training & Predicting data on KNN =3</h3>

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_predicted_3 = knn.predict(X_test)
```

<h3 style='color:blue' align="left">Training & Predicting data on KNN =5</h3>

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_predicted_5 = knn.predict(X_test)
```

<h3 style='color:blue' align="left">Training & Predicting data on KNN =7</h3>

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
y_predicted_7 = knn.predict(X_test)
```

<h3 style='color:blue' align="left">Training & Predicting data on KNN =9</h3>

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
y_predicted_9 = knn.predict(X_test)
```

<h3 style='color:blue' align="left">Training & Predicting data on KNN =11</h3>

```python
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train, y_train)
y_predicted_11 = knn.predict(X_test)
```

<h2 style='color:black' align="center">PART-II</h2>
<h4 style='color:black' align="center">Test the trained model on the data se provied in test.csv file</h4>

<h4 style='color:blue' align="Center">Testing with the predicted data for KNN=3. Confusion Matrix, Precision score, Recall score, and F1 score</h4>

```python
cm_3 = confusion_matrix(y_test, y_predicted_3)
```

```python
import seaborn as sn
plt.figure(figsize = (15,8))
sn.heatmap(cm_3, annot=True, fmt='0')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
ps = precision_score(y_test,y_predicted_3,average='micro')
print("Precision Score: {:.3f} for KNN=3".format(ps))
```

```python
rs = recall_score(y_test, y_predicted_3, average='micro')
print("Recall score: {:.3f} for KNN=3".format(rs))
```

```python
f1 = f1_score(y_test, y_predicted_3, average='micro')
print("F1 score: {:.3f} for KNN=3".format(f1))
```
